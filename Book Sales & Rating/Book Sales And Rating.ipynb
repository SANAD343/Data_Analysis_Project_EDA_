```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# DataSet Link: https://www.kaggle.com/datasets/thedevastator/books-sales-and-ratings

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
#  Get the Size of DataFrame
df.size
```

```python
# Get the Dimension of DataFrame
df.ndim
```

```python
# Get the type of each columns within the dataframe
df.dtypes
```

```python
# Stastical Summary

df.describe()
```

```python
# Fix the publishing year as it is having negative value

df=df[df['Publishing Year']>1900]
```

```python
df.describe()
```

```python
# Check for na data

df.isna().sum()
```

```python
# Drop the book name that have NA data

df.dropna(subset = 'Book Name')
```

```python
# Check for NA Value within the Dataframe
df.isna().sum()
```

```python
# Drop the NA Values
df.dropna(subset = 'Book Name', inplace =True)
```

```python
df.isna().sum()
```

```python
#  Check for duplicated data

df.duplicated().sum()
```

```python
# Check the unique entries

df.nunique()
```

```python
# Create a histogram for the publishing year

plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
# Create a Bar chart of book in each genre

df["genre"].value_counts()
```

```python
# Get the number of books in each genre 
df["genre"].value_counts().plot(kind = 'bar')
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
```

```python
# Perform the GroupBy 

df.groupby("Author")["Book_average_rating"].mean()
```

```python
type(df.groupby("Author")["Book_average_rating"].mean())
```

```python
df.groupby("Author")["Book_average_rating"].mean().reset_index()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
# Book ratings in each genre
sns.boxplot(x="genre", y = "Book_ratings_count", data = df, color = "red")
plt.xlabel("Genre")
plt.ylabel("Books Rating Count")
plt.title("Box Plot for book rating in each genre")
plt.show()
```

```python
# Scatter plot showing the relation between Unit sold and price of the book
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Unit Sold")
plt.title("Unit Sold Vs Price")
```

```python
# Get the total number of languages in which book is written

language_counts = df["language_code"].value_counts()
```

```python
language_counts
```

```python
# Represent the Distribution of books with respect to their language

plt.pie(language_counts,  labels= language_counts.index,
        startangle= 60, autopct="%1.1f%%")
plt.title('Language of Distribution of books')
plt.show()
```

```python
# Get the Publisher's revenue in Descending order

df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending= False)
```

```python
# get the max rating 

df.groupby('Author_Rating')['Book_ratings_count'].mean().max()
```

```python
df.groupby('language_code').size().sort_values(ascending=False)
```

```python
df.groupby('Author_Rating')['Book_ratings_count'].var()
```

```python
# Averege book rating vs Book Rating Count

plt.scatter(df['Book_average_rating'], df['Book_ratings_count'])
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.title('Scatter Plot of Book Average Rating vs Book Rating Count')
plt.show()
```

```python
# Get the total sales
total_gross_sales_by_author = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False)
```

```python
total_gross_sales_by_author
```

```python
total_gross_sales_by_author.plot(kind='bar')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total Gross Sales For Each Author')
plt.show()
```

```python
# Get the total sales for each author

total_gross_sales_by_author.head(20).plot(kind='bar')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total Gross Sales For Each Author')
plt.show()
```

```python
# Get the unit sold for each author rating
sns.boxplot(x='Author_Rating', y= 'units sold', data=df)
plt.xlabel('Author Rating')
plt.ylabel('Units Sold')
plt.title('Box Plot of Units Sold for Each Author Rating')
plt.show()
```

```python
# Get the total unit sold over the years
df.groupby('Publishing Year')['units sold'].sum().plot(kind='line', marker='o')
plt.xlabel('Publishing Year')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold Over the Years')
plt.show()
```
